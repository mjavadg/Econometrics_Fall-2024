## The problem is that, the years of months are not obvious

### Using conditions

```python
italy_core_inflation_rate = (4.21, 3.56, 3.08, 2.7, 2.33,
                             2.32, 2.1, 2, 1.9, 1.9, 1.9, 1.8) 
months = ("Oct", "Nov", "Dec", "Jan", "Feb", "Mar", 
          "Apr", "May", "Jun", "Jul", "Aug", "Sep")
index = 0

for rate, month in zip(italy_core_inflation_rate, months):
    if months[index] in ("Oct", "Nov", "Dec"):
        print(f"{index + 1}th period is {month.} 2023 inflation rate which is {rate}")
    else:
        print(f"{index + 1}th period is {month} 2024 inflation rate which is {rate}")
    index += 1
```

## Enumerate() when plotting

### different labels

```python
italy_core_inflation_rate = (4.21, 3.56, 3.08, 2.7, 2.33,
                             2.32, 2.1, 2, 1.9, 1.9, 1.9, 1.8) 
months = ("Oct", "Nov", "Dec", "Jan", "Feb", "Mar", 
          "Apr", "May", "Jun", "Jul", "Aug", "Sep")

plt.figure(figsize = (10, 6))
plt.plot(range(len(months)), italy_core_inflation_rate, marker = '*', linestyle = '--', color = 'magenta')

for index, rate in enumerate(italy_core_inflation_rate):
    if months[index] in ("Oct", "Nov", "Dec"):
        label = f"{months[index]} 2023"
    else: 
        label = f"{months[index]} 2024"
    plt.text(index, rate, label, fontsize = 10, ha = "left")
    
plt.title("Italy core inflation rate over 12 months")
plt.xlabel("Months")
plt.ylabel("Core inflation rate (%)")
plt.xticks(range(len(months)), months) #plt.xticks(ticks=None, labels=None, **kwargs)
plt.grid(True)
plt.show()
```

## Plotting functions

```python
import numpy as np

def f(K):
    e = K ** alpha
    return(e)
    
k = np.linspace(0, 100, 1000)
alpha = 0.5

a = f(k)

plt.figure(figsize = (10, 6))
plt.plot(a, color = "red", label = "f(k)")
plt.title("growth")
plt.legend()
plt.show()
```
