# plotting part 2

```python
A = np.linspace(1, 100, 100)
B = np.linspace(1, 50, 100)

plt.figure(figsize=(10, 6))
plt.plot(A, B, color = 'magenta', marker = "^", label = "Magenta") #first step
plt.title("A magenta colored line") #second step
plt.xlabel("A line") #third step
plt.ylabel("B line") #forth step
plt.xlim(1, 100)
plt.ylim(1, 50)
plt.grid(True) #fifth step
plt.legend() #to get labels printed in plt.plot() #sixth step
plt.show()
```

# Condtional Statements and Control Flow Structures

## If-Elif-Statements

```python
e = [1, 2, 3, 4]
f = [1, 5, 3, 9]

f == e
```

```python
e[0] == f[0]
```

```python
if e[0] == f[0]:
    print("the first items are equal")
```

```python
if e[0] == f[0]:
    print("the first items are equal")
```

## Nested for loop

```python
for i in e:
    for j in f:
        if i == j:
            print(i, j, "equal items")
    else: print(i, j, "unequal items")
    
```
