## Preliminary concepts

### Fields

Let *F* denote either the set of real numbers of the set of complex numbers, then:

1. Addition is commutative:

$$ x + y = y + x $$

for all x and y in F.

2. Addition is associative:

$$ x + ( y + z) = (x + y) + z $$

for all x, y, and z in F.

3. There is a unique element 0 (zero) in F such that x + 0 = x, for every x in F.

4. To each x in F there corresponds a unique element (-x) in F such that x + (-x) = 0

5. Multiplication is commutative:

$$ xy = yx $$

for all x and y in F.

6. Multiplication is associative:

$$ x(yz) = (xy)z$$

for all x, y, and z in F

7. There is a unique non-zero element 1 (one) in F such that x1 = x for every x in F.

8. To each non-zero x in F there corresponds a unique element x^-1 (or 1/x) in F such that xx^-1 = 1.

9. Multiplication distributes over addition; that is, x(y + z) = xy + xz, for all x, y, and z in F.

### Common Examples of Fields:

- Set of rational numbers, real numbers, and complex numbers

### Sub-field

- A **sub-field** is subset of a field that is itself under the same operations. For example, the set of rational numbers is a sub-field of the field of real numbers.

### Vector Space or Linear Space

consists of the following:

1. a field F of scalars;

2. a set **V** of objects, called vectors;

3. a rule (or operation), called vector addition, which associates with each pair of vectors α, β in **V** a vector. α + β in **V**, called the sum of α and β, in such a way that:

(a) addition is commutative, α + β = β + α

(b) addition is associative, α + (β + γ) = (α + β) + γ

(c) there is a unique vector in 0 in **V**, called the zero vector, such that α + 0 = α for all α in **V**;

(d) for each vector α in **V** there is unique vector -a in **V** such that α + (-α) = 0;

4. a rule (or operation), called scalar mulitiplication, which associates with each scalar c in F and vector α in **V**. A vector *cα* in **V**, called the product of c and α, in such a way that

(a) $$1α = α \text{ for every } α \text{ in } V;$$

(b) $$(c_{1} c_{2})α = c_{1}(c_{2}α);$$

(c) $$c(α + β) = cα + cβ;$$

(d) $$(c_{1} + c_{2})α = c_{1}α + c_{2}α.$$

**In short: A vector space is a collection of vectors that can be added together and multiplied by scalars. For example, the space $ \mathbb{R}^n $ is the set of all n-dimensional vectors.**

### Subspace: 
- A subspace is a subset of a vector space that is also a vector space. For example, the set of all solutions to a homogeneous linear equation forms a subspace of $ \mathbb{R}^n $.

## Graphical examples of a 2d vector space and a 1d subspace

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Define the vector space (2D plane)
x = np.linspace(-5, 5, 100)

# Plot the vector space (the entire plane)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect(aspect = 2, adjustable='datalim')

#aspect = "auto" or "equal"
#adjustable='box'


# Draw the vector space (the full plane)
ax.fill_between(x, -5, 5, color='lightblue', alpha=0.8, 
                label='Vector Space (R^2)')

# Define a subspace (a line through the origin)
subspace_x = np.linspace(-5, 5, 100)
subspace_y = 2 * subspace_x  # Example line y = 2x

# Plot the subspace
ax.plot(subspace_x, subspace_y, color='red', label='Subspace (Line through origin)')

ax.axhline(0, color='black', linewidth=0.5, ls='-')
ax.axvline(0, color='black', linewidth=0.5, ls='-')
ax.grid(color='gray', linestyle='--', linewidth=0.5)
ax.legend()
ax.set_title('Vector Space and Subspace')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

## The plot consists of three parts:

- fig = plt.figure() which creates the chart

- ax = fig.add_subplot(111, projection = '3d') which creates the 3d plot

- ax.plot_surface(X, Y, Z) which prepares variable to be plotted

a breakdown of what 111 means:

The first digit (1) represents the number of rows in the grid. The second digit (1) represents the number of columns in the grid. The third digit (1) represents the index of the subplot in the grid. So, 111 means that you are creating a grid with 1 row and 1 column, and you are placing the subplot in the first (and only) position in that grid. Essentially, it creates a single subplot that takes up the entire figure.

If you had a grid with more subplots, you could use different numbers to specify their positions. For example, fig.add_subplot(221) would create a 2x2 grid and place the subplot in the first position.

```python
#defining the coub douglas utility function
def cobb_douglas(x, y):
    alpha = 0.3
    beta = 0.7
    return (x ** alpha) * (y ** beta)

#define a the ranges for good x and good y
X = np.linspace(1, 100, 100)
Y = np.linspace(1, 100, 100)
X, Y = np.meshgrid(X, Y)

#calculate the utility
U = cobb_douglas(X, Y)

#creating the plot
fig = plt.figure(figsize = (12, 6))

#3d
ax1 = fig.add_subplot(121, projection = '3d')
ax1.plot_surface(X, Y, U, cmap = 'viridis')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('utility level')
ax.set_title("utility function")

#contour plot for indifference curve
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, U, levels = 10, cmap = 'autumn')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax.set_title("indifference curves")
plt.colorbar(contour, ax = ax2, label = "Utility")


plt.tight_layout()
plt.show()
```

## Applying the same to the system of equations below

\begin{align*}
6x + 5y + 2z &= 49 \quad (1) \\
3x + 4y + z &= 32 \quad (2) \\
x + 2y + 3z &= 20 \quad (3)
\end{align*}

```python
import numpy as np

# Coefficients matrix and constants for the 3D system
A = np.array([[6, 5, 2], [3, 4, 1], [1, 2, 3]])  # 3x3 coefficients matrix
B = np.array([49, 32, 20])                      # Constants

# Calculate the solution
solutions = np.linalg.solve(A, B)
print("Solutions using numpy:", solutions)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define a grid for plotting the planes
x = np.linspace(-10, 10, 10)
y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(x, y)

# Define Z for each plane based on the equations
Z1 = (49 - 6*X - 5*Y) / 2  # From equation 1: 6x + 5y + 2z = 49
Z2 = (32 - 3*X - 4*Y)      # From equation 2: 3x + 4y + z = 32
Z3 = (20 - X - 2*Y) / 3    # From equation 3: x + 2y + 3z = 20

# Plot the planes
surface1 = ax.plot_surface(X, Y, Z1, alpha=0.4, color='blue',
                           rstride=100, cstride=100,
                           label='Plane 1: 6x + 5y + 2z = 49')

surface2 = ax.plot_surface(X, Y, Z2, alpha=0.2, color='orange',
                           rstride=100, cstride=100,
                           label='Plane 2: 3x + 4y + z = 32')

surface3 = ax.plot_surface(X, Y, Z3, alpha=0.6, color='green',
                           rstride=100, cstride=100, 
                           label='Plane 3: x + 2y + 3z = 20')

# Manually set facecolors and edgecolors
surface1._facecolors2d = surface1._facecolor3d
surface1._edgecolors2d = surface1._edgecolor3d
surface2._facecolors2d = surface2._facecolor3d
surface2._edgecolors2d = surface2._edgecolor3d
surface3._facecolors2d = surface3._facecolor3d
surface3._edgecolors2d = surface3._edgecolor3d

# Highlight the solution point
ax.scatter(solutions[0], solutions[1], solutions[2], color='red', s=100, 
           label='Solution Point')
ax.text(solutions[0], solutions[1], solutions[2], f'Solution: {solutions}',
        fontsize=10, ha='right')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot of the Linear System of Equations')

# Set limits for better visualization
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])

# Add legend
ax.legend()

# Show the plot
plt.show()
```

- Dimension of solution space = Number of variable - Rank

if dimension of solution space is 0 it means we have unique solution to the system.

## So the A matrix in this system is full rank. Now let's have a look at a system that is linearly dependent.

\begin{align*} 
6x + 5y + 2z &= 49 \quad (1) \\
3x + 4y + z &= 32 \quad (2) \\
x + \frac{5}{6}y + \frac{1}{3}z &= \frac{49}{6} \quad (3) 
\end{align*}

```python
import numpy as np

# Coefficients matrix and constants for the 3D system
A2 = np.array([[6, 5, 2], [3, 4, 1], [1, 5/6, 1/3]])  # 3x3 coefficients matrix
B2 = np.array([49, 32, 49/6])                         # Constants

# Check if the matrix is singular
if np.linalg.det(A2) != 0:
    solutionss = np.linalg.solve(A2, B2)
    print("Solutions using numpy:", solutionss)
else:
    solutionss = None
    print("The matrix is singular and cannot be solved.")

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define a grid for plotting the planes
x = np.linspace(-10, 10, 10)
y = np.linspace(-10, 10, 10)
X, Y = np.meshgrid(x, y)

# Define Z for each plane based on the equations
Z1 = (49 - 6*X - 5*Y) / 2  # From equation 1: 6x + 5y + 2z = 49
Z2 = (32 - 3*X - 4*Y)      # From equation 2: 3x + 4y + z = 32
Z3 = (49/6 - X - (5/6)*Y) / (1/3)  # From equation 3: x + 5/6y + 1/3z = 49/6

# Plot the planes
surface1 = ax.plot_surface(X, Y, Z1, alpha=0.5, color='magenta', 
                           rstride=100, cstride=100, label='Plane 1: 6x + 5y + 2z = 49')

surface2 = ax.plot_surface(X, Y, Z2, alpha=0.5, color='black', 
                           rstride=100, cstride=100, label='Plane 2: 3x + 4y + z = 32')

surface3 = ax.plot_surface(X, Y, Z3, alpha=0.5, color='green', 
                           rstride=100, cstride=100, label='Plane 3: x + 5/6y + 1/3z = 49/6')

# Manually set facecolors and edgecolors
surface1._facecolors2d = surface1._facecolor3d
surface1._edgecolors2d = surface1._edgecolor3d
surface2._facecolors2d = surface2._facecolor3d
surface2._edgecolors2d = surface2._edgecolor3d
surface3._facecolors2d = surface3._facecolor3d
surface3._edgecolors2d = surface3._edgecolor3d

# Highlight the solution point if it exists
if solutionss is not None:
    ax.scatter(solutions[0], solutionss[1], solutionss[2],
               color='red', s=100, label='Solution Point')
    ax.text(solutionss[0], solutionss[1], solutionss[2],
            f'Solution: {solutions}', fontsize=10, ha='right')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot of the Linear System of Equations')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])
ax.legend()
plt.show()
```

| Number of Variables | Rank       | Dimension of Solution Space | Implication                          |
|---------------------|------------|-----------------------------|--------------------------------------|
| \( n \)             | \( n \)    | 0                           | Unique solution                      |
| \( n \)             | \( r < n \)| \( n - r > 0 \)            | Infinite solutions (free variables)  |
| \( n \)             | \( r > n \)| Negative                    | No solution (inconsistent system)    |


    Unique Solution: When the rank equals the number of variables.
    Infinite Solutions: When the rank is less than the number of variables.
    No Solution: When the rank exceeds the number of variables.
