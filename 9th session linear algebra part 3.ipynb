The set of all scalar multiples of a nonzero vector **V** is a straight line through the origin, here is denoted by $\mathcal{L}[\mathbf{v}]$ and call it a line generated or spnnaed by **V**.  For example if **V** = (1,0), then $\mathcal{L}[\mathbf{v}]$ is X axis in ${R}^n$. If **V** = (1,1) in ${R}^n$, then $\mathcal{L}[\mathbf{v}]$ is the diagonal line in figure below.

```python
import matplotlib.pyplot as plt
import numpy as np

#define the vector
v = np.array([1,1])

x = np.linspace(-100, 100, 100)
y = x * (v[1]/v[0])

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y,label = "L(v)", color = "purple")
ax.axhline(0, color = 'red', linewidth = 0.5, ls = '-')
ax.axvline(0, color = 'red', linewidth = 0.5, ls = '-')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_xlabel("x_1")
ax.set_ylabel("x_2")
ax.set_title("The line L(v) spanned by vector v")
ax.legend()

#add an arrow
arrow_color = 'red'
aa_color = 'magenta'
arrow_thickness = 2
ax.arrow(0,0,1,1, head_width = 0.1, head_length = 0.2, fc = arrow_color, ec = aa_color, linewidth = arrow_thickness)

plt.show()
```

## Definition 1:

- A vector β in **V** is said to be a linear combination of the vectors $α_{1}, ..., α_{n}$ in **V** provided there exist scalars $c_{1}, ..., c_{n}$ in F such that 

$$ 
β = c_{1}α_{1}+ ..., c_{n}α_{n} = \sum_{i=1}^{n} c_{i}\alpha_{i}
$$

## Definition 2:

- We can take all possible linear combinations of $\mathbf{v_{1}}$ and $\mathbf{v_{2}}$ to obtain the set spanned by $\mathbf{v_{1}}$ and $\mathbf{v_{2}}$:

$$\mathcal{L}[\mathbf{v_1}, \mathbf{v_2}] \equiv {c_1}\mathbf{v_1} + {c_2}\mathbf{v_2} : c_1 \in \mathbb{R} \text{ and } c_2 \in \mathbb{R}$$

If $\mathbf{v_{2}}$ is multiple of $\mathbf{v_{1}}$, then $\mathcal{L}[\mathbf{v_1}, \mathbf{v_2}] = \mathcal{L}[\mathbf{v_1}]$ as in figure below

```python
#define the vector
v1 = np.array([1,2])
v2 = np.array([2, 4])

#generating the first line
x = np.linspace(-100, 100, 100)
y1 = x * (v1[1]/v1[0])

#generating the second line
y2 = x * (v2[1]/v2[0])

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y1,label = "L(v_1)", color = "purple")
ax.plot(x, y2,label = "L(v_2)", color = "pink")
ax.axhline(0, color = 'red', linewidth = 0.1, ls = '-')
ax.axvline(0, color = 'red', linewidth = 0.1, ls = '-')
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_xlabel("x_1")
ax.set_ylabel("x_2")
ax.set_title("The lines L(v_1) and L(v_2) spanned by vector v")
ax.legend()

#add arrow1
arrow_color = 'red'
aa_color = 'magenta'
arrow_thickness = 2
ax.arrow(0,0,1,2, head_width = 0.1, head_length = 0.2, fc = arrow_color, ec = aa_color, linewidth = arrow_thickness)
ax.text(0.5, 0.5, "v1")

#add arrow2
arrow_color = 'black'
aa_color = 'black'
arrow_thickness = 2
ax.arrow(0,0,2,4, head_width = 0.1, head_length = 0.2, fc = arrow_color, ec = aa_color, linewidth = arrow_thickness)
ax.text(2.5, 2.5, "v2")

plt.show()
```

- If $\mathbf{v_{2}}$ is multiple of $\mathbf{v_{1}}$, or vice versa, we say that If $\mathbf{v_{2}}$ and $\mathbf{v_{1}}$ are *linearly dependent*. Otherwise, we say that they are linearly independent. 

If $\mathbf{v_{2}}$ is multiple of $\mathbf{v_{1}}$, we write

$$ \mathbf{v_{2}} = c_{1}\mathbf{v_{1}} \text{ or } \mathbf{v_{2}} -c_{1}\mathbf{v_{1}} = 0 $$

same goes for when $\mathbf{v_{1}}$ is multiple of $\mathbf{v_{2}}$.

# Combining the two definitions

- If  $\mathbf{v_{1}}$ and $\mathbf{v_{2}}$ are **linearly dependent** then there exist scalars $c_{1}$ and $c_{2}$, *not both zero*, so that

$$ c_{1}\mathbf{v_{1}} + c_{2}\mathbf{v_{2}} = \mathbf{0}, c_{1} \text{ or } c_{2} \text{ nonzero }$$ 

- If the two vectors are **linearly independent** if:
$$ c_{1}\mathbf{v_{1}} + c_{2}\mathbf{v_{2}} = \mathbf{0} \implies c_{1} \text{ or } c_{2} \text{ nonzero }$$

### Expanding:

$$\mathcal{L}[\mathbf{v_1}, \mathbf{v_2}, \mathbf{v_3}] \equiv {c_1}\mathbf{v_1} + {c_2}\mathbf{v_2} + {c_3}\mathbf{v_3}: c_1, c_2, c_3 \in \mathbb{R}$$

# Definiton 3:

- Vectors $\mathbf{v_{1}}, \mathbf{v_{2}},..., \mathbf{v_{k}}$ in $\mathbb{R}^n$ are **linearly dependent** if and only if there exist scalars $c_{1},c_{2},...,c_{k}$, *not all zero*, such that

$$ c_1\mathbf{v_{1}} + c_2\mathbf{v_{2}} + ... + c_k\mathbf{v_{k}} = \mathbf{0}$$

- Vectors $\mathbf{v_{1}}, \mathbf{v_{2}},..., \mathbf{v_{k}}$ in $\mathbb{R}^n$ are **linearly independent** if and only if $ c_1\mathbf{v_{1}} + c_2\mathbf{v_{2}} + ... + c_k\mathbf{v_{k}} = \mathbf{0}$ for scalars $c_{1},c_{2},...,c_{k}$ implies that $c_{1} = c_{2} = ... = c_{k}=0$

## Span of a Vector Space

- Linear combinations of $\mathbf{v_1}, ..., \mathbf{v_k}$:

$$\mathcal{L}[\mathbf{v_1}, ..., \mathbf{v_k}] \equiv {c_1}\mathbf{v_1} + ... + {c_k}\mathbf{v_k}: c_1, ..., c_3 \in \mathbb{R}$$

- The set $\mathcal{L}[\mathbf{v_1}, ..., \mathbf{v_k}]$ is generated or spanned by $\mathbf{v_1}, ..., \mathbf{v_k}$.


- If evey vector in vector space **V** can be written as a linear comination of $\mathbf{v_1}, ..., \mathbf{v_k}$ **span** V.:

$$ V = \mathcal{L}[\mathbf{v_1}, ..., \mathbf{v_k}]$$ 

We say that $\mathbf{v_1}, ..., \mathbf{v_k}$ **span** V.

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

u = np.linspace(-5, 5, 100)
v = np.linspace(-5, 5, 100)
U, V = np.meshgrid(u, v)

X = U * v1[0] + V * v2[0]
Y = U * v1[1] + V * v2[1]
Z = U * v1[2] + V * v2[1]

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X, Y, Z, alpha = 0.7, color = 'lightgray')

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color = 'red', linewidth = 2, arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color = 'red', linewidth = 2, arrow_length_ratio=0.1)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title('span of the vector space')

plt.show()
```

###  Using the standard basis vectors, we can span the entire space, meaning any vector in this space can be represented as a linear combination of these basis vectors.

```python
import numpy as np

# Define the standard basis vectors as 3x1 matrices
e1 = np.array([[1], [0], [0]])  # Standard basis vector e1
e2 = np.array([[0], [1], [0]])  # Standard basis vector e2
e3 = np.array([[0], [0], [1]])  # Standard basis vector e3

# Combine them into a 3x3 matrix
standard_basis = np.hstack((e1, e2, e3))

# Print the matrix in a column-aligned format
print("\nStandard basis of 3D vector space")
for row in standard_basis:
    print(f"{row[0]}  {row[1]}  {row[2]}")
```

```python
e1 = np.array([1, 0, 0])
e2 = np.array([0, 1, 0])
e3 = np.array([0, 0, 1])

def travel_by_standard_basis(n1, n2, n3):
    r1 = n1 * e1
    r2 = n2 * e2
    r3 = n3 * e3
    return r1 + r2 + r3
```

## Linear Transformation

## Definition: 

A **linear transformation** is a function $T: V \to W$ between two vector spaces $V$ and $W$ that satisfies two properties:
  - **Additivity**: $T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + (\mathbf{v})$ for all $(\mathbf{u}, \mathbf{v} \in V)$
  - **Homogeneity**: $T(c\mathbf{u}) = cT(\mathbf{u}))$ for all $(\mathbf{u} \in V)$ and scalar $(c)$

- **Matrix Representation**: Any linear transformation can be represented by a matrix \(A\) such that:

$$
T(\mathbf{x}) = A\mathbf{x}
$$

where $(\mathbf{x})$ is a vector in $(V)$.

Image and Range: The image (or range) of a linear transformation $T$ is the set of all vectors that can be written as $T(\mathbf{v})$ for some $\mathbf{v}$ \in V. It represents the output space of the transformation.

Dimension and Rank: The dimension of the image of $T$ is called the rank of $T$. It indicates the number of linearly independent vectors in the image.

## A linear transformation preserves vector addition and scalar multiplication.

## Example 1

Consider a linear transformation $T: \mathbb{R}^2 \to \mathbb{R}^2$ defined by the matrix:

$$
A = \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix}
$$


- Let the basis for $\mathbb{R}^2$ be $\{\begin{bmatrix} 1 \\ 0 \end{bmatrix}, \begin{bmatrix} 0 \\ 1 \end{bmatrix}\}$.
- Applying the transformation:

```python
A = np.array([[2, 0], [0, 3]])
basis_vectors = np.array([[1,0], [0, 1]])

# Apply the transformation to the basis vectors
transformed_vectors = A @ basis_vectors

print("original basis vectors: ")
print(basis_vectors)
print("\ntransformed vectors")
print(transformed_vectors)
```

### In previous sessions, we solved systems of linear equations using Cramer’s rule and the np.linalg.solve() function. Another method is by using the row reduced echelon form.

Consider the following system of equations:

$$
\begin{align*}
2x + y - z &= 1 \\
4x - 2y + 5z &= 2 \\
-2x + 3y + 2z &= 3
\end{align*}
$$

- Write the augmented matrix for this system.

$$
\begin{bmatrix}
2 & 1 & -1 & | & 1 \\
4 & -2 & 5 & | & 2 \\
-2 & 3 & 2 & | & 3
\end{bmatrix}
$$

- Use row reduction to determine if the system has a solution. If so, find the solution.


```python
import sympy as sp

# Define the augmented matrix
A = sp.Matrix([[2, 1, -1, 1],
                [4, -2, 5, 2],
                [-2, 3, 2, 3]])

# Perform row reduction
RREF, pivot_columns = A.rref()

# Display the Reduced Row Echelon Form
print("Reduced Row Echelon Form:")
print(RREF)
```
